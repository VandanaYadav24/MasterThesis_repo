```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv("datasets/preparation_of_datasets/Amazon_Music/Preprocessed_Amazon_Music_dataset.csv")
```

```python
df2
```

```python
df2[(df2["label"]>=4) & (df2["label"]<=5)]
```

```python
df2[df2["label"]==3]
```

```python
p = df2.pivot(columns='label', values='label').plot.hist(legend=False)
plt.xticks(df2['label'].unique())
plt.xlabel('Ratings',fontsize=12, fontname="Times New Roman")
plt.ylabel('Number of reviews',fontsize=12, fontname="Times New Roman")
plt.title('Distribution of ratings of Amazon Digital Music dataset', fontsize=15, fontname="Times New Roman")
p.get_figure().savefig("pics/Amazon_Music_distribution_of_ratings.png", dpi=500)
```
